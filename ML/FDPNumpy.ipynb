```python
import numpy as np
```

```python
# Create 1-D Array 
arr = np.array([10, 20, 30, 40,])
print(arr)
print(np.mean(arr)) # print the average us function np.mean()
```

```python
A = np.zeros((3,3)) # 3 x 3 directory with zeros
print(A)
```

```python
A = np.ones((2,4)) # 2 rows and 4 column
print(A)
```

```python
A = np.full((2,3), 4) # 2 rows and 3 columns, all valus 4
print(A)
```

```python
A = np.arange(1, 10, 2) # skip 2 element arrange(start, end, skip) start is include but end is excluded
print(A)
```

```python
A = np.arange(0, 2, 4) # stop value less than skip then only start value is taken
print(A)
```

```python
A = np.eye(3) 
A
```

```python
A = np.linspace(0,10,5) # Five evenly space number fro 0 to 10, included both 0 and 10
print(A)
```

```python
#list of random numbers between 0 and 1
A = np.random.rand(10)
print(A)
```

```python
# random numbres for 2-D Array
A = np.random.rand(3,2)
print(A)
```

```python
# random number not in between 0 and 1
A = np.random.randint(10, 100, size = (2,3))
print(A)
```

```python
# random number not in between 0 and 1, Array of 5 integers
A = np.random.randint(10, 100, size = (5))
print(A)
print(np.mean(A)) # find average of newly created array
```
